```python
def preaga(cols):
    Age=cols[0]
    Pclass=cols[1]
    
    
    if pd.isnull(Age):
        if Pclass == 1:
            return 37
        elif Pclass == 2:
            return 29
        else:
            return 24
    
    else:
        return Age
```

def col(Name):
    if 'Mrs' in Name:
        return 'Mrs'
    elif 'Mr' in Name:
        return 'Mr'
    elif 'Miss' in Name:
        return 'Miss'
    elif 'Lady' in Name:
        return 'Miss'
    elif 'Dr' in Name:
        return 'Mr'
    elif 'Rev' in Name:
        return 'Mr'
    elif 'Major' in Name:
        return 'Mr'
    elif 'Master' in Name:
        return 'Master'
    elif 'countess' in Name:
        return 'Mrs'
    elif 'Ms' in Name:
        return 'Miss'
    else:
        return 'Mr'

train['Name']=train['Name'].apply(col)

```python
from sklearn.preprocessing import LabelEncoder

def label_encode_columns(df, columns):
    encoder = LabelEncoder()
    for col in columns:
        df[col] = encoder.fit_transform(df[col])
    return df
```

```python
df=label_encode_columns(train[['Name', 'Sex','Embarked' ]],['Name', 'Sex','Embarked' ])

train=train.drop(['Name', 'Sex','Embarked','Ticket'],axis=1)

train=pd.concat([train,df],axis=1)

train=train.drop('PassengerId',axis=1)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaled_values = scaler.fit_transform(train[['Pclass', 'Age', 'Parch', 'Fare', 'Sex',
       'Embarked']])
train[['Pclass', 'Age', 'Parch', 'Fare','Sex',
       'Embarked']] = scaled_values
```

```python
import numpy as np
import tensorflow as tf
```

```python
X = train.drop('Survived', axis=1)  # Features
y = train['Survived']               # Labels

```

```python
X=train.drop('Survived',axis=1)
y= train['Survived']

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)

from sklearn.linear_model import LogisticRegression

lg = LogisticRegression()

lg.fit(X_train,y_train)

prediction=lg.predict(X_test)

from sklearn.metrics import classification_report

print(classification_report(y_test,prediction))



print(classification_report(y_test,prediction))

from sklearn.metrics import confusion_matrix

print(confusion_matrix(y_test,prediction))
```

```python
#X_train = X_train.reshape(-1, 28*28).astype("float32") / 255.0
#X_test = X_test.reshape(-1, 28*28).astype("float32") / 255.0

# One-hot encode the labels
y_train = tf.keras.utils.to_categorical(y_train, 2)
y_test = tf.keras.utils.to_categorical(y_test, 2)
```

```python
input_size = 6
output_size = 1
# Use same hidden layer size for both hidden layers. Not a necessity.
hidden_layer_size = 50
tf.keras.utils.set_random_seed(42)
model = tf.keras.Sequential([
    tf.keras.layers.InputLayer(input_shape=(input_size,)),
    tf.keras.layers.Dense(hidden_layer_size,  activation= tf.keras.activations.relu),
    tf.keras.layers.Dense(hidden_layer_size, activation='relu'),
    tf.keras.layers.Dense(hidden_layer_size, activation='relu'),
    tf.keras.layers.Dense(hidden_layer_size, activation='relu'),
    tf.keras.layers.Dense(hidden_layer_size, activation='relu'),
    tf.keras.layers.Dense(output_size,activation='sigmoid')  # No activation because we'll apply softmax inside loss
])

# Compile the model
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001),
    loss=tf.keras.losses.binary_crossentropy,
    metrics=['accuracy']
)

# Training
batch_size = 20
max_epochs = 200

history = model.fit(
    X_train, y_train,
    validation_data=(X_test, y_test),
    batch_size=batch_size,
    epochs=max_epochs,
    callbacks=[
        tf.keras.callbacks.EarlyStopping(patience=2, restore_best_weights=True)
    ]
)
```
